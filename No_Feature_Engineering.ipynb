```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('Customer_Behaviour.csv')
data
```

```python
data.info()
```

# Preprocessing

```python
def preprocess_input(df):
    df = df.copy()
    # Drop the User ID column
    df = df.drop('User ID', axis=1)
    return df
```

```python
X = preprocess_input(data)
X
```

```python
X['Gender'].unique()
```

```python
X = preprocess_input(data)  # Existing preprocessing
X['Gender'] = X['Gender'].map({'Female': 0, 'Male': 1})  # Binary encoding
X
```

```python
# Split df into x and y
y = X['Purchased']
X = X.drop('Purchased', axis=1)
#train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=1)
```

```python
print("X shape:", X.shape)
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
X_train
```

```python
print('=====this is X_train=====')
display(X_train)
print('=====this is X_test=====')
display(X_test)
print('=====this is y_train=====')
display(y_train)
print('=====this is X_test=====')
display(y_test)
```

```python
# Scale X
scaler = StandardScaler()
scaler.fit(X_train)
X_train = pd.DataFrame(scaler.transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
```

```python
display(X_train)
display(X_test)
```

# Training & Results

```python
model = LogisticRegression()
model.fit(X_train, y_train)

acc = model.score(X_test, y_test)
print('Accuracy: {:.3f}'.format(acc*100))
```
